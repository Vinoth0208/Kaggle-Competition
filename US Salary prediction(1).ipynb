```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

data
model = LabelEncoder()
for col in ['ID', 'Jobs_Group', 'Profile', 'Company', 'City', 'State',
       'Frecuency_Salary', 'Skills', 'Sector_Group', 'Revenue']:
    data[col] = model.fit_transform(data[col])

X = data.drop(['Mean_Salary'],axis=1)
y = df['Mean_Salary']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

modelpredict = LinearRegression()
modelpredict.fit(x_train,y_train)
train_pred = modelpredict.predict(x_train)
test_pred = modelpredict.predict(x_test)
print("Train error: ",mean_absolute_percentage_error(y_train,train_pred))
print("Test error: ",mean_absolute_percentage_error(y_test,test_pred))
```
